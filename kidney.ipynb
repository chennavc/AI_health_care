```python
for x in df.columns:
    if(df[x].dtypes) == "object":
        print(x,df[x].unique())
```

```python
df["cad"]=df["cad"].replace('\tno', 'no')
```

```python
for x in df.columns:
    if(df[x].dtypes) == "object":
        print(x,df[x].unique())
```

```python
# import labelencoder
from sklearn.preprocessing import LabelEncoder
# instantiate labelencoder object
le = LabelEncoder()
# Categorical boolean mask
categorical_feature_mask = df.dtypes==object
# filter categorical columns using mask and turn it into a list
categorical_cols = df.columns[categorical_feature_mask].tolist()
```

```python
# apply le on categorical feature columns
df[categorical_cols] = df[categorical_cols].apply(lambda col: le.fit_transform(col))
df[categorical_cols]
```

```python
#binary lable encoding manually
df["pc"]=df["pc"].replace('normal', 0)
df["pc"]=df["pc"].replace('abnormal', 1)
df["pcc"]=df["pcc"].replace('notpresent', 0)
df["pcc"]=df["pcc"].replace('present', 1)
df["ba"]=df["ba"].replace('notpresent', 0)
df["ba"]=df["ba"].replace('present', 1)
df["htn"]=df["htn"].replace('no', 0)
df["htn"]=df["htn"].replace('yes', 1)
df["dm"]=df["dm"].replace('no', 0)
df["dm"]=df["dm"].replace('yes', 1)
df["pe"]=df["pe"].replace('no', 0)
df["pe"]=df["pe"].replace('yes', 1)
df["cad"]=df["cad"].replace('no', 0)
df["cad"]=df["cad"].replace('yes', 1)
df["ane"]=df["ane"].replace('no', 0)
df["ane"]=df["ane"].replace('yes', 1)
df["appet"]=df["appet"].replace('poor', 0)
df["appet"]=df["appet"].replace('good', 1)
df["classification"]=df["classification"].replace('notckd', 0)
df["classification"]=df["classification"].replace('ckd', 1)


```

```python
# Compute the correlation matrix
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(X,y)
```

```python
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
```

```python
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(5).plot(kind='barh')
plt.show()
feat_importances.nlargest(5)
```

```python
df1=pd.DataFrame()
df1["htn"]=df["htn"]
df1["sg"]=df["sg"]
df1["dm"]=df["dm"]
df1["al"]=df["al"]
df1["hemo"]=df["hemo"]
```

```python
X=df1
y=y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train=X_train.values
y_train=y_train.values
X_test=X_test.values
y_test=y_test.values
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
# fit model no training data
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
